FIBBONACCI SERIES

```python

def fib(n):
    a = 0
    b = 1
    print(a)
    print(b)
    for i in range(2,n):
        c = a+b
        a = b
        b = c
        print(c)    
fib(10)
    
```

 Factorial

```python
def fact(n):
    f = 1
    for i in range(1,n+1):
        f = f*i
    return f    
result = fact(5)
print(result)
```

Factorial with Recursion

```python
def fact(n):
    if (n == 1):
        return 1
    else:
        return n*fact(n-1)
result = fact(5)
print(result)
```

Anonymous Function : Function without name

```python
mul = lambda a : a*a
result = mul(4)
print(result)
```

```python
nums  = [1,5,4,7,8,9,6,2]
even = list(filter(lambda n : n%2 == 0,nums))
odd = list(filter(lambda n : n%2 != 0,nums))
print(even)
print(odd)
```

```python
double  = list(map(lambda a : a*3,even))
print(double)
```

```python
from functools import reduce
sum = reduce(lambda a,b : a+b ,double)
print(sum)
```

Decorators

```python
def div(a,b):
    return a/b
def smart_div(func):
    
    def inner(a,b):
        if a<b :
            a,b = b,a
        return func(a,b)
    return inner
div = smart_div(div)
div(2,4)
```
