```python
def new():
        inpt = input("Enter : ")
        print("Hello",current_thread().getName())
def new2():
        inpt = input("Enter : ")
        print("Hello",current_thread().getName())
t1 = Thread(target = new)
t2 = Thread(target = new2)
print(current_thread().getName())
t1.start()
t1.join()
t2.start()
t2.join()
```

# By Extending The Thread Class

```python
import threading
class A(threading.Thread): 
    def run(self): 
        for i in range(7): 
            print("Hi",current_thread().getName())

obj = A() 
obj.start() 
obj.join() 
print("Hello",current_thread().getName())
```

# Without Extending the Thread Class

```python
class ex:
    def C(self):
        lst = [2,4,5,'kausik','abc']
        for i in lst:
            print(i)
            
my_obj = ex()
t1 = Thread(target = my_obj.C)
t1.start()
t1.join()
print("Done")
            
```
