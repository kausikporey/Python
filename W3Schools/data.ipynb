```python
import pandas as pd
df = pd.read_csv('dirtydata.csv')
print(df)
```

```python
df['Calories'].fillna(130,inplace=True)
print(df)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(df.to_string())
```

```python
df.dropna(subset=['Date'],inplace=True)
print(df)
```

```python
for x in df.index:
    if df.loc[x,'Duration']>120:
        df.drop(x,inplace=True)
print(df)        
```

```python
df.drop_duplicates()
```
